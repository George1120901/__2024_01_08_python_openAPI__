```python
# å®šç¾©å‡½æ•¸
def number_sum_limit(nums, limit):
    result = []
    s = 0
    for num in nums:
        if len(result) == limit:
            break
        if type(num) == str:
            continue
        result.append(num)
        s += num
    
    return result, 'sum = {}'.format(s)
# å‘¼å«å‡½æ•¸
number_sum_limit(nums = [7.5, 8, 'aa', 4.75, 'bb', 2, 8, 5, 6], limit = 5)
```

## while loops

ä¸€èˆ¬è€Œè¨€ï¼Œfor-loop æ¯”è¼ƒé©ç”¨åœ¨ã€Œå·²çŸ¥è¿´åœˆæ•¸ã€çš„å•é¡Œï¼Œè€Œ while-loop å‰‡é©ç”¨åœ¨ã€Œç„¡æ³•é çŸ¥è¿´åœˆæ•¸ã€çš„å•é¡Œä¸Šã€‚

ç•¶ç¨‹å¼éœ€è¦ä¸æ–·åœ°é‡è¦†æŸäº›é‹ç®—ï¼Œä¸€ç›´åˆ°å‡ºç¾æŒ‡å®šçš„ç‰¹æ®Šç‹€æ³æ™‚æ‰åœæ­¢ï¼Œé€™ç¨®æƒ…å½¢å°±æ¯”è¼ƒé©åˆç”¨ while-loop ä¾†å¯¦ç¾ã€‚

while-loop åœ¨åŸ·è¡Œå…¶å…§éƒ¨æ•˜è¿°ä¹‹å‰ï¼Œéƒ½æœƒå…ˆæª¢æŸ¥æ¢ä»¶åˆ¤æ–·å¼ï¼Œå¦‚æœå…¶åˆ¤æ–·å€¼ç‚ºçœŸ (True)ï¼Œå‰‡åŸ·è¡Œè¿´åœˆå…§éƒ¨çš„æ•˜è¿°ï¼›å¦å‰‡å°±çµæŸ while-loopã€‚

ä¾‹å¦‚å…©å€‹äººç©çŒœæ•¸å­—éŠæˆ²ï¼Œç”²æ–¹å¿ƒä¸­é»˜æƒ³ä¸€å€‹æ•¸è®“ä¹™æ–¹å»çŒœï¼Œç”²æ–¹å°±ä¹™æ–¹æ‰€çŒœçš„æ•¸å­—å£é ­æç¤ºã€Œå†å¤§ä¸€é»ã€æˆ–ã€Œå†å°ä¸€é»ã€ï¼Œç›´åˆ°ä¹™æ–¹çŒœä¸­ç”²æ–¹å¿ƒä¸­é»˜æƒ³çš„ç‰¹å®šæ•¸å­—æ‰åœæ­¢ï¼Œé€™æ¨£çš„éŠæˆ²å°±å±¬æ–¼ç„¡æ³•é çŸ¥è¿´åœˆæ•¸çš„å•é¡Œï¼Œç›¸è¼ƒæ–¼ for-loop ï¼Œæ­¤å•é¡Œæœƒæ›´é©åˆä½¿ç”¨ while-loop ä¾†å¯¦ç¾ã€‚

while-loop çš„èªæ³•å¦‚ä¸‹ï¼š
>>>while test_expression:
>>>   Body of while

```python
#è‹¥æˆ‘å€‘è¦ç”¢ç”Ÿä¸€ç­† 1, 2, 3, â€¦, 10 çš„åºåˆ—ï¼Œç”¨ while-loop å¯ä»¥é€™æ¨£å¯«ï¼š

i = 1
while i <= 10:
    print("\t",i, end=" ")
    i = i + 1
```

```python
# åˆ©ç”¨ while-loop è¨ˆç®— 1 åŠ åˆ° 10 çš„å’Œï¼Œä¸¦å°‡æ¯æ¬¡ç´¯åŠ ä¹‹å¾Œçš„å€¼åˆ—å°å‡ºä¾†ï¼š

sum = 0
i = 1
while i <= 10:
    sum += i     # sum = sum + i
    i += 1       # i = i + 1
print(sum, end= " ")
```

```python
#ä¹Ÿå¯åˆ©ç”¨ break æ•˜è¿°ä¾†ä¸­æ–· while-loop çš„åŸ·è¡Œã€‚
#æœ¬ä¾‹ï¼Œwhile å¾Œé¢çš„åˆ¤æ–·å€¼ç‚º 1ï¼Œä»£è¡¨è¿´åœˆæ°¸é æˆç«‹ï¼Œæœƒä¸€ç›´åŸ·è¡Œä¸‹å»ã€‚ä½†æˆ‘å€‘åœ¨è¿´åœˆå…§éƒ¨åŠ ä¸Š if æ•˜è¿°çš„åˆ¤æ–·æ¢ä»¶å’Œ break æ•˜è¿°ï¼Œè®“è¿´åœˆå¯ä»¥åœ¨ç‰¹å®šæƒ…æ³ä¸‹çµæŸåŸ·è¡Œã€‚ä¾‹å¦‚ï¼š

sequences = [0, 1, 2, 3, 4, 5]
i = 0
while 1: #åˆ¤æ–·æ¢ä»¶å€¼ç‚º1, ä»£è¡¨è¿´åœˆæ°¸é æˆç«‹
    print(sequences[i], end = " ")
    i += 1 #i = i + 1
    if i == len(sequences):
        print("\t","No elements left.")
        break
```

```python
# Example 1 : åˆ—å° 1, 2, 3, â€¦, 10çš„åºåˆ—ï¼Œä½†åˆ©ç”¨continueè·³é 4 é€™å€‹æ•¸å­—ï¼š

i = 0
while i < 10:
    i += 1      # i = i + 1
    if i == 4:
        continue
    print (i, end = " ")
```

```python
# Example 2 : åˆ—å° 9, 8, 7, â€¦, 0çš„åºåˆ—ï¼Œä½†åˆ©ç”¨continueè·³é 4 é€™å€‹æ•¸å­—ï¼š

i = 10
while i > 0:
    i -= 1  # i = i - 1
    if i == 4:
        continue
    print(i, end = " ")
```
