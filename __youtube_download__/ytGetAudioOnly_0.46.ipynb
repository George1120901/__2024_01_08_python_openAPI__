```python
# V0.4.0 -->完成 (2024.06.15止-->final)
from pytube import YouTube
import subprocess
import os

def media_contents(url):
    # 影片格式分析
    print('標題: ',yt.title)                 # 影片標題
    print('長度 ( 秒 ): ',yt.length)         # 影片長度 ( 秒 )
    print('作者: ',yt.author)                # 影片作者
    print('作者頻道網址: ',yt.channel_url)   # 影片作者頻道網址
    print('縮圖網址: ',yt.thumbnail_url)     # 影片縮圖網址
    print('觀看數: ',yt.views)               # 影片觀看數

    # 分析
    #print(yt.streams)  # 影片支援哪些畫質
    print('stream數: ',len(yt.streams))

    for st in yt.streams:
        print(st)
    print('media格式分析完成')


def download_video_and_audio(url,itag_Vedio,itag_Audio):
    print('downdloading of video & audio .....')
    itag_List = [itag_Vedio,itag_Audio]
    for i in itag_List:
        stream = yt.streams.get_by_itag(i)
        stream.download(output_path='C:/Users/Lu/Documents/youtube',filename=f'python_{yt.title}_itag({i}).mp4', skip_existing=True, max_retries=0)
        print(f'python_{yt.title}_itag({i})完成')
    print('downdload完成')


def merge_video_and_audio(video_path, audio_path, output_path):
    command = [
        'ffmpeg',
        '-i', video_path,
        '-i', audio_path,
        '-c', 'copy',
        '-map', '0:v',
        '-map', '1:a',
        output_path
    ]
    
    try:
        print("執行merge命令:", ' '.join(command))  # 增加執行內容訊息
        result = subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, encoding='utf-8',timeout=30)
        print("合併成功！輸出文件位於:", output_path)
    except subprocess.CalledProcessError as e:
        print("ffmpeg合併過程中發生錯誤:", e)
        if e.stderr:
            try:
                # 嘗試用 UTF-8 解碼
                print("ffmpeg命令輸出:", e.stderr)
            except UnicodeDecodeError:
                # 如果失敗，使用 CP950 解碼
                print("ffmpeg命令輸出:", e.stderr.decode(encoding='cp950', errors='ignore'))
        else:
            print("無法獲取 ffmpeg 錯誤輸出。")
    except subprocess.TimeoutExpired as e:
        print("命令執行超時:", e)


#url = input("Enter the url of media : ")   
url='https://youtube.com/live/o6HQepxxDlo'   # 2023_11_10_下午
yt = YouTube(url) 
itag_Audio = 140
itag_Vedio = 137 
media_contents(url)
download_video_and_audio(url,itag_Vedio,itag_Audio)
merge_video_and_audio(f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Vedio}).mp4', f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Audio}).mp4', f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag(merge).mp4')

os.remove(f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Vedio}).mp4')
os.remove(f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Audio}).mp4')
print('合併成功！影像&音訊檔案刪除完成')

#<2023_11_10_下午> media 長度 03:10:30 容量大小731,905KB 處理耗時10m 6.4s
```

### 設定時間限制(timeout)之討論

#### 透過chatGPT -1
##### Question : 若無預期的錯誤，但很久無法完成，如何印出訊息，跳出執行？
##### Answer:
要處理在執行下載過程中遇到長時間無法完成的情況，可以使用 signal 模組來設定超時處理。在超時發生時，會拋出一個自定義的例外，從而中斷下載過程並打印錯誤訊息。以下是示範程式碼：

```python
# for iOS ???????
import signal
from pytube import YouTube
from pytube.exceptions import PytubeError

# 自定義的超時例外
class TimeoutException(Exception):
    pass

# 處理超時的函數
def timeout_handler(signum, frame):
    raise TimeoutException()

# 設定超時（例如 60 秒）
TIMEOUT = 60

def download_video(video_url):
    # 在開始下載前設定超時
    signal.signal(signal.SIGALRM, timeout_handler)
    signal.alarm(TIMEOUT)
    
    try:
        video = YouTube(video_url)
        stream = video.streams.get_highest_resolution()
        stream.download()
        print("下載成功！")
    except TimeoutException:
        print("下載過程超時，請重試。")
    except PytubeError as e:
        print("發生 pytube 錯誤:", e)
    except Exception as e:
        print("發生其他錯誤:", e)
    finally:
        # 取消超時
        signal.alarm(0)

if __name__ == "__main__":
    video_url = input("輸入 YouTube 視頻網址: ")
    download_video(video_url)
```

在這段程式碼中，我們做了以下幾點：

定義了一個自定義的例外 TimeoutException。
使用 signal 模組來設定超時處理函數 timeout_handler。
在開始下載前，設定了一個 60 秒的超時。
在 try 區塊中進行視頻下載操作。如果下載過程中超過 60 秒未完成，則會觸發 TimeoutException 並打印相應的錯誤訊息。
在 finally 區塊中取消超時設定，以防止影響後續操作。
這樣，如果下載過程超過設定的時間限制，程式將中斷執行並打印提示訊息。

#### 透過chatGPT -2
##### Question : stream.download(timeout=60)無法運用嗎？
##### Answer :
pytube 庫中的 download 方法本身並沒有內建的超時機制。因此，無法直接通過設置 timeout 參數來控制下載時間。不過，可以通過一些額外的處理來實現類似的功能。

使用 signal 模組設置超時是比較常見的方法之一，但需要注意的是，這種方法在 Windows 平台上可能會遇到一些問題，因為 signal.alarm 在 Windows 上不受支持。

如果你需要在 Windows 上實現下載超時，可以考慮使用多線程或多進程來達到相似的效果。例如，使用 threading 模組來實現超時控制。以下是一個示例：

```python
# for windows
import threading
import time
from pytube import YouTube
from pytube.exceptions import PytubeError

# 自定義的超時例外
class TimeoutException(Exception):
    pass

def download_video_with_timeout(video_url, timeout):
    def target():
        try:
            video = YouTube(video_url)
            stream = video.streams.get_highest_resolution()
            stream.download()
            print("下載成功！")
        except PytubeError as e:
            print("發生 pytube 錯誤:", e)
        except Exception as e:
            print("發生其他錯誤:", e)

    thread = threading.Thread(target=target)
    thread.start()
    thread.join(timeout)
    
    if thread.is_alive():
        print("下載過程超時，請重試。")
        raise TimeoutException()

if __name__ == "__main__":
    video_url = input("輸入 YouTube 視頻網址: ")
    timeout = 60  # 設置超時秒數
    try:
        download_video_with_timeout(video_url, timeout)
    except TimeoutException:
        print("下載操作被強制終止。")
```

在這個示例中：

定義了 download_video_with_timeout 函數，其中啟動了一個新線程來執行下載操作。
使用 thread.join(timeout) 方法等待下載線程完成，並設置超時時間。如果超時，則打印提示訊息並拋出 TimeoutException。
在主程序中，捕獲 TimeoutException 並打印相應的提示訊息。
這種方法在 Windows 和 Unix 系統上都能正常工作，可以有效地設置下載操作的超時時間。

#### 透過chatGPT -3
##### Question : thread = threading.Thread(target=target)語法為何？
##### Answer :
threading.Thread(target=target) 是用來創建一個新的線程（Thread）。在這個語法中：
1. threading.Thread 是 Python 標準庫中的 threading 模組的一個類，用來創建並管理線程。
2. target=target 是指定線程要執行的目標函數。在這個例子中，target 是一個函數名。

以下是對這個語法的詳細解釋：
1. threading.Thread 類：
  threading.Thread 是用來創建一個新的線程對象。每個線程對象都會運行指定的目標函數。
2. target=target：
  target 參數用來指定線程在啟動後要執行的函數。這個函數不接受任何參數。
  在這個例子中，target 是一個函數名，這個函數會在新線程中運行。

以下是一個完整的例子，說明如何使用 threading.Thread 創建並啟動一個新線程來執行指定的函數：

```python
import threading
import time

# 定義一個函數，將在線程中運行
def print_hello():
    for i in range(5):
        print(f"Hello from thread {i}")
        time.sleep(1)  # 每次打印後等待1秒

# 創建一個線程對象，將 `print_hello` 函數作為目標函數
thread = threading.Thread(target=print_hello)

# 啟動線程，開始執行 `print_hello` 函數
thread.start()

# 在主線程中也打印一些信息
for i in range(5):
    print(f"Hello from main thread {i}")
    time.sleep(0.5)  # 每次打印後等待0.5秒

# 等待子線程結束
thread.join()

print("Thread finished.")
```

在這個例子中：

創建了一個名為 print_hello 的函數，這個函數將在新線程中運行。
使用 threading.Thread(target=print_hello) 創建了一個新的線程對象，並將 print_hello 函數指定為目標函數。
使用 thread.start() 啟動新線程，這會開始執行 print_hello 函數。
在主線程中，打印一些信息來顯示主線程和子線程是如何並行運行的。
使用 thread.join() 等待子線程結束。
這樣，新線程和主線程可以同時運行，這就是多線程編程的基本原理。

#### 透過chatGPT -4
##### Question : thread.join()一定需要嗎？
##### Answer :
thread.join() 在多線程編程中並不是一定需要的，但它在許多情況下是有用的。以下是一些情況以及是否需要 thread.join() 的原因：

##### 何時需要使用 thread.join()：
1.等待線程完成：
  如果你需要確保某個線程完成它的工作後，才能繼續執行主程序中的後續步驟，那麼你應該使用 thread.join()。
例如，如果你有一個線程負責下載文件，而主程序需要處理這些文件，那麼你需要確保下載完成後再進行處理。這時可以使用 thread.join()。
2.同步多個線程：
  當你有多個線程，並且需要等待所有線程完成後再進行下一步操作，可以對每個線程調用 join()。
這樣可以確保主程序在所有線程結束後再繼續執行。

##### 何時不需要使用 thread.join()：
1.非阻塞執行：
  如果你希望主程序和線程並行執行，而不需要等待線程完成，那麼可以不使用 thread.join()。
例如，在 GUI 程序中，你可能希望在背景線程中執行長時間運行的任務，而不阻塞主線程的用戶界面更新。
2.獨立任務：
  如果線程執行的任務與主程序完全獨立，沒有相互依賴，也可以不使用 thread.join()。
例如，記錄日誌的線程可以獨立運行，無需等待它結束。

示例代碼
這裡有兩個示例，分別說明使用和不使用 thread.join() 的情況：

使用 thread.join()

```python
# 使用 thread.join()
import threading
import time

def worker():
    print("Worker thread started")
    time.sleep(2)
    print("Worker thread finished")

# 創建並啟動線程
thread = threading.Thread(target=worker)
thread.start()

# 等待線程結束
thread.join()
print("Main thread finished after worker thread")
```

在這個示例中，主程序會等待 worker 線程完成後再繼續執行，保證 Worker thread finished 先於 Main thread finished after worker thread 被打印。

不使用 thread.join()

```python
# 不使用 thread.join()
import threading
import time

def worker():
    print("Worker thread started")
    time.sleep(2)
    print("Worker thread finished")

# 創建並啟動線程
thread = threading.Thread(target=worker)
thread.start()

# 主程序不等待線程結束
print("Main thread finished immediately")
```

在這個示例中，主程序不等待 worker 線程完成，會立即打印 Main thread finished immediately，然後 worker 線程會在其自己的時間內完成。

總結來說，是否使用 thread.join() 取決於具體需求。如果需要等待線程完成才能繼續下一步操作，那麼應該使用 join()；如果不需要等待，則可以不使用。

### 使用FFmpeg將音頻MP4轉換為MP3?之討論

#### 透過chatGPT -1
##### Question :使用FFmpeg將音頻MP4轉換為MP3?
##### Answer:
要處理在執行下載過程中遇到長時間無法完成的情況，可以使用 signal 模組來設定超時處理。在超時發生時，會拋出一個自定義的例外，從而中斷下載過程並打印錯誤訊息。以下是示範程式碼：
要使用 FFmpeg 將音頻從 MP4 轉換為 MP3，您可以使用以下命令：

>>>ffmpeg -i input.mp4 -vn -acodec libmp3lame -q:a 2 output.mp3

這裡是這個命令的詳細說明：
1. ffmpeg: 調用 FFmpeg 工具。
2. -i input.mp4: 指定輸入文件 input.mp4。
3. -vn: 禁用視頻，因為我們只需要音頻部分。
4. -acodec libmp3lame: 指定 MP3 編碼器。libmp3lame 是一個高質量的 MP3 編碼器。
5. -q:a 2: 設置音頻質量。q:a 選項允許設置 VBR（可變比特率）的質量級別，數字 2 表示非常高的質量（範圍從 0 到 9，0 為最佳）。
6. output.mp3: 指定輸出文件 output.mp3。

您可以將此命令用於命令行中，也可以在 Python 程序中通過 subprocess 模組調用它。這裡有一個示例 Python 程序，演示如何使用 subprocess 來運行 FFmpeg 命令：

```python
import subprocess

def convert_mp4_to_mp3(input_file, output_file):
    command = [
        'ffmpeg',
        '-i', input_file,
        '-vn',
        '-acodec', 'libmp3lame',
        '-q:a', '2',
        output_file
    ]
    
    try:
        result = subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
        print(f"Conversion successful! MP3 file is located at: {output_file}")
    except subprocess.CalledProcessError as e:
        print("An error occurred during conversion:", e)
        if e.stderr:
            print("FFmpeg error output:", e.stderr)
        else:
            print("Could not retrieve FFmpeg error output.")

# Example usage
convert_mp4_to_mp3("input.mp4", "output.mp3")
```

這個程序調用 convert_mp4_to_mp3 函數，將 input.mp4 轉換為 output.mp3。如果轉換成功，它會打印成功消息；如果發生錯誤，則會打印錯誤消息和 FFmpeg 的錯誤輸出。

#### 透過chatGPT -2
##### Question : 當abr=128kbps 時，-q:a 要設置為何?
##### Answer:
在FFmpeg中，-q:a 和 -b:a 是用來設定音頻質量和比特率的參數。這兩者通常不會同時使用。-q:a（質量）用來控制VBR（可變比特率）的音質，數值範圍從0到9，數值越小音質越好。而 -b:a 用來設定CBR（固定比特率）的音頻比特率。

如果你想要設定音頻比特率為128kbps，應該使用 -b:a 而不是 -q:a。

以下是使用 -b:a 的正確方式

```python
def convert_mp4_to_mp3(input_file, output_file):
    command = [
        'ffmpeg',
        '-i', input_file,
        '-vn',
        '-acodec', 'libmp3lame',
        '-b:a', '128k',
        output_file
    ]
    try:
        result = subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
        print("轉換成功！輸出文件位於:", output_file)
    except subprocess.CalledProcessError as e:
        print("ffmpeg轉換過程中發生錯誤:", e)
        if e.stderr:
            print("ffmpeg命令輸出:", e.stderr)
        else:
            print("無法獲取 ffmpeg 錯誤輸出。")

# 範例使用
convert_mp4_to_mp3("input.mp4", "output.mp3")
```

-q:a 0 大約是 320 kbps
-q:a 1 大約是 256 kbps
-q:a 2 大約是 192 kbps
-q:a 3 大約是 160 kbps
-q:a 4 大約是 128 kbps
-q:a 5 大約是 96 kbps
所以如果你希望音質接近128kbps，可以設定 -q:a 4。但請注意這只是近似值。

解釋
\w 匹配字母、數字和下劃線，即 [a-zA-Z0-9_]
\u4e00-\u9fff 匹配 Unicode 范圍內的中文字符
[\w\u4e00-\u9fff]+ 匹配一個或多個中英文字符
如果你想要更精確地匹配，可能需要根據實際需求進行調整。例如，如果你只想匹配中英文字符而不包括數字，可以使用 [a-zA-Z\u4e00-\u9fff]+。

更多示例
只匹配中英文字符（不包括數字和下劃線）：
>>>pattern = r'[a-zA-Z\u4e00-\u9fff]+'

匹配中英文字符及空格（允許空格分隔的詞組）：
>>>pattern = r'[a-zA-Z\u4e00-\u9fff\s]+'

匹配中英文字符及標點符號：
>>>pattern = r'[a-zA-Z\u4e00-\u9fff.,!?，。！？]+'

使用這些模式可以靈活匹配各種包含中英文字符的字符串。根據具體需求選擇合適的正則表達式模式。
