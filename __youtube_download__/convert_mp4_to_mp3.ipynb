```python
# 2024.07.11  V0.96 --> OK (v0.9 final) 
# V0.96 --> copy of V0.95_2
# convert_mp4_to_mp3(input_file, output_file) : '-q:a', '2',   -->預設
# convert_mp4_to_mp3重複(同檔名.mp3)對應 -3 --> open 後,close
# file_path 追加 & 簡化
# 調整 os.chdir(f'{output_path}')執行時序

from pytube import YouTube
import re
import subprocess
import os

def streams_analysis(url):
    yt = YouTube(url) 
    # 影片格式分析
    print(f'標題: {yt.title}')              # 影片標題
    print(f'長度(s): {yt.length}')          # 影片長度 ( 秒 )
    print(f'作者: {yt.author}')             # 影片作者
    print(f'網址: {yt.channel_url}')        # 影片作者頻道網址
    print(f'縮圖網址: {yt.thumbnail_url}')  # 影片縮圖網址
    print(f'觀看數: {yt.views}')            # 影片觀看數

    # 分析
    print(f'支援畫質:{yt.streams}')  # 影片支援哪些畫質
    print('影片支援畫質數: ',len(yt.streams))
    for st in yt.streams:
        print(st)
    print('media格式分析完成')

def dir_path(output_path):
    try:
        os.chdir(f'{output_path}')  # 解決:os.mkdir如果目錄已經存在，FileExistsError則會引發 --> OK
        print(f'資料夾 {output_path} existed')
    except:
        #os.mkdir(f'{os.path.dirname(filepath)}')
        os.makedirs(f'{output_path}')
        print(f'建立新檔案資料夾 {output_path}')
        os.chdir(f'{output_path}')

    print(f'預設檔案位置 {output_path}')
    print(f'目前資料夾{os.getcwd()}') 

def get_audio_only(url):
    yt = YouTube(url)   
    print('全名 : ',yt.title)
    list=['/',':','"','<','>','\?','\*'] # window檔案格式排除符號
    yt_titles = yt.title
    for i in list:
        yt_title = re.sub(f'{i}','_',yt_titles)
        yt_titles = yt_title
    print(yt_titles)

    #yt_title = re.search(r'[A-Za-z0-9_ ]+',yt_titles).group()
    yt_title = re.search(r'[\w\u4e00-\u9fff\s]+',yt_titles).group() #設定包含中英文字
    print('簡稱(檔名) : ',yt_title)

    yt_audio = yt.streams.filter(abr='128kbps').first()
    yt_audio.download(output_path=output_path,filename=f'yt_audio_{yt_title}.mp4', skip_existing=True, max_retries=0)  
    print(f'downdload < {yt.title} > 完成')
    return yt_title

def convert_mp4_to_mp3(input_file, output_file):
    command = [
        'ffmpeg',
        '-i', input_file,
        '-vn',
        '-acodec', 'libmp3lame',
        '-q:a', '2',
        output_file
    ]
    

    try:
        f = open(f'{output_path}/yt_audio_{yt_title}.mp3','r',encoding= 'utf-8')  # convert_mp4_to_mp3重複(同檔名.mp3)對應 -3
        print(f'{output_path}/yt_audio_{yt_title}.mp3 已存在')
        f.close()
    except:    
        try:
            result = subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True,encoding= 'utf-8')  # 設定encoding='utf-8'
            print("轉換成功！輸出文件位於:", output_file)
        except subprocess.CalledProcessError as e:
            print("ffmpeg轉換過程中發生錯誤:", e)
            if e.stderr:
                print("ffmpeg命令輸出:", e.stderr)
            else:
                    print("無法獲取 ffmpeg 錯誤輸出。")
   

# url = 'https://www.youtube.com/watch?v=sWdGVa4vz5E'  # 曲目:蔡琴,Tsai Chin,情人的眼泪,明日之歌,飘零的落花,月光小夜曲,70后80后90后最喜欢的华语金曲,Appreciation of Chinese songs

work_item = int(input('輸入 ( 1 :分析   2:下載   其他:跳出執行 )'))
if work_item == 1:
   url = str(input("輸入影音網址 (Enter the url of media) : ")) 
   streams_analysis(url) 
elif work_item == 2:
    url = str(input("輸入影音網址(Enter the url of media) : "))  
    dir_name = 'youtube'  # 設定工作/儲存檔案之資料夾
    file_path = f'c:/{dir_name}/filename.ext'
    output_path = os.path.dirname(file_path)
    print(f'out_path >>> {output_path}') 
    dir_path(output_path)
    yt_title = get_audio_only(url)  # 返回結果方式
    convert_mp4_to_mp3(f'{output_path}/yt_audio_{yt_title}.mp4', f'{output_path}/yt_audio_{yt_title}.mp3')
else:
     print('輸入 ( 1 :分析   2:下載   其他:跳出執行 )')


# media 長度 01:06:05 mp4容量大小62,678KB mp3容量大小88,762KB 處理耗時1m 45.2s
```
