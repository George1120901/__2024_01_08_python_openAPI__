```python
# V0.4.0 -->å®Œæˆ (2024.06.15æ­¢-->final)
from pytube import YouTube
import subprocess
import os

def media_contents(url):
    # å½±ç‰‡æ ¼å¼åˆ†æ
    print('æ¨™é¡Œ: ',yt.title)                 # å½±ç‰‡æ¨™é¡Œ
    print('é•·åº¦ ( ç§’ ): ',yt.length)         # å½±ç‰‡é•·åº¦ ( ç§’ )
    print('ä½œè€…: ',yt.author)                # å½±ç‰‡ä½œè€…
    print('ä½œè€…é »é“ç¶²å€: ',yt.channel_url)   # å½±ç‰‡ä½œè€…é »é“ç¶²å€
    print('ç¸®åœ–ç¶²å€: ',yt.thumbnail_url)     # å½±ç‰‡ç¸®åœ–ç¶²å€
    print('è§€çœ‹æ•¸: ',yt.views)               # å½±ç‰‡è§€çœ‹æ•¸

    # åˆ†æ
    #print(yt.streams)  # å½±ç‰‡æ”¯æ´å“ªäº›ç•«è³ª
    print('streamæ•¸: ',len(yt.streams))

    for st in yt.streams:
        print(st)
    print('mediaæ ¼å¼åˆ†æå®Œæˆ')


def download_video_and_audio(url,itag_Vedio,itag_Audio):
    print('downdloading of video & audio .....')
    itag_List = [itag_Vedio,itag_Audio]
    for i in itag_List:
        stream = yt.streams.get_by_itag(i)
        stream.download(output_path='C:/Users/Lu/Documents/youtube',filename=f'python_{yt.title}_itag({i}).mp4', skip_existing=True, max_retries=0)
        print(f'python_{yt.title}_itag({i})å®Œæˆ')
    print('downdloadå®Œæˆ')


def merge_video_and_audio(video_path, audio_path, output_path):
    command = [
        'ffmpeg',
        '-i', video_path,
        '-i', audio_path,
        '-c', 'copy',
        '-map', '0:v',
        '-map', '1:a',
        output_path
    ]
    
    try:
        print("åŸ·è¡Œmergeå‘½ä»¤:", ' '.join(command))  # å¢åŠ åŸ·è¡Œå…§å®¹è¨Šæ¯
        result = subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, encoding='utf-8',timeout=30)
        print("åˆä½µæˆåŠŸï¼è¼¸å‡ºæ–‡ä»¶ä½æ–¼:", output_path)
    except subprocess.CalledProcessError as e:
        print("ffmpegåˆä½µéç¨‹ä¸­ç™¼ç”ŸéŒ¯èª¤:", e)
        if e.stderr:
            try:
                # å˜—è©¦ç”¨ UTF-8 è§£ç¢¼
                print("ffmpegå‘½ä»¤è¼¸å‡º:", e.stderr)
            except UnicodeDecodeError:
                # å¦‚æœå¤±æ•—ï¼Œä½¿ç”¨ CP950 è§£ç¢¼
                print("ffmpegå‘½ä»¤è¼¸å‡º:", e.stderr.decode(encoding='cp950', errors='ignore'))
        else:
            print("ç„¡æ³•ç²å– ffmpeg éŒ¯èª¤è¼¸å‡ºã€‚")
    except subprocess.TimeoutExpired as e:
        print("å‘½ä»¤åŸ·è¡Œè¶…æ™‚:", e)


#url = input("Enter the url of media : ")   
url='https://youtube.com/live/o6HQepxxDlo'   # 2023_11_10_ä¸‹åˆ
yt = YouTube(url) 
itag_Audio = 140
itag_Vedio = 137 
media_contents(url)
download_video_and_audio(url,itag_Vedio,itag_Audio)
merge_video_and_audio(f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Vedio}).mp4', f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Audio}).mp4', f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag(merge).mp4')

os.remove(f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Vedio}).mp4')
os.remove(f'C:/Users/Lu/Documents/youtube/python_{yt.title}_itag({itag_Audio}).mp4')
print('åˆä½µæˆåŠŸï¼å½±åƒ&éŸ³è¨Šæª”æ¡ˆåˆªé™¤å®Œæˆ')

#<2023_11_10_ä¸‹åˆ> media é•·åº¦ 03:10:30 å®¹é‡å¤§å°731,905KB è™•ç†è€—æ™‚10m 6.4s
```

```python
# for iOS ???????
import signal
from pytube import YouTube
from pytube.exceptions import PytubeError

# è‡ªå®šç¾©çš„è¶…æ™‚ä¾‹å¤–
class TimeoutException(Exception):
    pass

# è™•ç†è¶…æ™‚çš„å‡½æ•¸
def timeout_handler(signum, frame):
    raise TimeoutException()

# è¨­å®šè¶…æ™‚ï¼ˆä¾‹å¦‚ 60 ç§’ï¼‰
TIMEOUT = 60

def download_video(video_url):
    # åœ¨é–‹å§‹ä¸‹è¼‰å‰è¨­å®šè¶…æ™‚
    signal.signal(signal.SIGALRM, timeout_handler)
    signal.alarm(TIMEOUT)
    
    try:
        video = YouTube(video_url)
        stream = video.streams.get_highest_resolution()
        stream.download()
        print("ä¸‹è¼‰æˆåŠŸï¼")
    except TimeoutException:
        print("ä¸‹è¼‰éç¨‹è¶…æ™‚ï¼Œè«‹é‡è©¦ã€‚")
    except PytubeError as e:
        print("ç™¼ç”Ÿ pytube éŒ¯èª¤:", e)
    except Exception as e:
        print("ç™¼ç”Ÿå…¶ä»–éŒ¯èª¤:", e)
    finally:
        # å–æ¶ˆè¶…æ™‚
        signal.alarm(0)

if __name__ == "__main__":
    video_url = input("è¼¸å…¥ YouTube è¦–é »ç¶²å€: ")
    download_video(video_url)
```

```python
# for windows
import threading
import time
from pytube import YouTube
from pytube.exceptions import PytubeError

# è‡ªå®šç¾©çš„è¶…æ™‚ä¾‹å¤–
class TimeoutException(Exception):
    pass

def download_video_with_timeout(video_url, timeout):
    def target():
        try:
            video = YouTube(video_url)
            stream = video.streams.get_highest_resolution()
            stream.download()
            print("ä¸‹è¼‰æˆåŠŸï¼")
        except PytubeError as e:
            print("ç™¼ç”Ÿ pytube éŒ¯èª¤:", e)
        except Exception as e:
            print("ç™¼ç”Ÿå…¶ä»–éŒ¯èª¤:", e)

    thread = threading.Thread(target=target)
    thread.start()
    thread.join(timeout)
    
    if thread.is_alive():
        print("ä¸‹è¼‰éç¨‹è¶…æ™‚ï¼Œè«‹é‡è©¦ã€‚")
        raise TimeoutException()

if __name__ == "__main__":
    video_url = input("è¼¸å…¥ YouTube è¦–é »ç¶²å€: ")
    timeout = 60  # è¨­ç½®è¶…æ™‚ç§’æ•¸
    try:
        download_video_with_timeout(video_url, timeout)
    except TimeoutException:
        print("ä¸‹è¼‰æ“ä½œè¢«å¼·åˆ¶çµ‚æ­¢ã€‚")
```
